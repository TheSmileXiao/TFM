```python
def normalize(segmentation, width, height):
    for i in range(0, len(segmentation), 2):
        segmentation[i] = segmentation[i] / width
        segmentation[i + 1] = segmentation[i + 1] / height
```

```python
# shutil.rmtree('./coco_annotation')
# os.mkdir('./coco_annotation')

basedir = './coco_annotation'
f = open('./validation/labels.json')
data = json.load(f)
txt_names = []
img_ids = []
for img in data['images']:
    img_base_name = img['file_name'].split('.')[0]
    txt_name = img_base_name + '.txt'
    txt_names.append(txt_name)
    img_ids.append(img['id'])
    open(f'{basedir}/{img_base_name}.txt', 'w')

img_dic = {}
for img_id, img_name in zip(img_ids, txt_names):
    img_dic[img_id] = img_name

for annontation in data['annotations']:
    img_id =  annontation['image_id']
    file_name = img_dic.get(img_id)
    file_dir = os.path.join(basedir, file_name)

    for img_info in data['images']:
        if img_info['id'] == img_id:
            width = img_info['width']
            height = img_info['height']
            break

    category_id = correct_category(annontation['category_id'])

    segmentation = annontation['segmentation']

    with open(file_dir, 'a') as f:
        for seg in segmentation:
            if isinstance(seg, list):
                normalize(seg, width = width, height = height)
                seg = ' '.join(map(str, seg))
                f.write(f'{category_id} {seg}\n')
```
